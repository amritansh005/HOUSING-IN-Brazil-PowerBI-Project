<h2>Converting different sheets from the main excel file into seperate excel files.</h2>

```python
# Load the entire Excel file, with each sheet in a dictionary
file_path = '/content/drive/MyDrive/Datasets of WorldQuant University/Housing in mexico/010-housing-in-mexico.xlsx'
xls = pd.ExcelFile(file_path)

# Iterate through each sheet and save it as a new Excel file
for sheet_name in xls.sheet_names:
    # Load each sheet into a DataFrame
    df = pd.read_excel(xls, sheet_name=sheet_name)

    # Save each DataFrame to a new Excel file named after the sheet
    output_path = f"/content/drive/MyDrive/Datasets of WorldQuant University/Housing in mexico/{sheet_name}.xlsx"
    df.to_excel(output_path, index=False)

    print(f"Saved {sheet_name}.xlsx at {output_path}")
```

<h2>Converting different excel files that were created in the previous step from the main excel file into seperate csv files for further use.</h2>

```python
# Define the directory containing the Excel files
directory_path = '/content/drive/MyDrive/Datasets of WorldQuant University/Housing in mexico/'

# List of specific Excel files to convert
excel_files_to_convert = [
    'brasil-real-estate-1.xlsx',
    'brasil-real-estate-2.xlsx',
    'mexico-real-estate-1.xlsx',
    'mexico-real-estate-2.xlsx',
    'mexico-real-estate-3.xlsx'
]

# Iterate through the list of specified Excel files
for filename in excel_files_to_convert:
    # Construct the full path to the Excel file
    file_path = os.path.join(directory_path, filename)

    # Check if the file exists in the directory
    if os.path.exists(file_path):
        # Load the Excel file into a DataFrame
        df = pd.read_excel(file_path)

        # Define the output CSV file path (same name as Excel but with .csv extension)
        csv_output_path = os.path.join(directory_path, f"{os.path.splitext(filename)[0]}.csv")

        # Save the DataFrame as a CSV file
        df.to_csv(csv_output_path, index=False)

        print(f"Converted {filename} to {csv_output_path}")
    else:
        print(f"File {filename} not found in the directory.")
```

```python
def extract_second_element(location_string):
  """
  Extracts the second element from a pipe-delimited string.

  Args:
    location_string: A string with elements separated by pipes (|).

  Returns:
    The second element in the string, or None if it doesn't exist.
  """
  try:
    return location_string.split('|')[2]
  except IndexError:
    return None

# Apply the function to the 'place_with_parent_names' column
df1['state'] = df1['place_with_parent_names'].apply(extract_second_element)

# Print the unique values of the new column
print(df1['state'].unique())
```

```python
df1.drop(columns=['place_with_parent_names'],inplace=True)
```

```python
# Split the 'lat-lon' column into two separate columns
df1[['lat', 'lon']] = df1['lat-lon'].str.split(',', expand=True)
```

```python
df1.drop(columns=['lat-lon'],inplace=True)
```

```python
# Remove '$' and ',' from the 'price_usd' column
df1['price_usd'] = df1['price_usd'].astype(str).str.replace(r'[$,]', '', regex=True)

# Convert the 'price_usd' column to numeric
df1['price_usd'] = pd.to_numeric(df1['price_usd'])
```

```python
# Convert 'area_m2' to float
df1['area_m2'] = df1['area_m2'].astype(float)
df1['lat'] = df1['lat'].astype(float)
df1['lon'] = df1['lon'].astype(float)
```

```python
df1.dropna(inplace=True)
```

```python
# Conversion rate
usd_to_brl_rate = 5.86873

# Calculate price_usd
df2['price_usd'] = df2['price_brl'] / usd_to_brl_rate
```

```python
df2.drop(columns=['price_brl'],inplace=True)
```

```python
df2.dropna(inplace=True)
```
